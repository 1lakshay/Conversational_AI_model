```python
import torch
from trl import SFTTrainer
from peft import LoraConfig
from datasets import load_dataset
from transformers import (AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, TrainingArguments, pipeline)
```

```python
import PyPDF2
```

```python
file = "corporate_laws.pdf"
pdf_reader = PyPDF2.PdfReader(file)
no_of_pages = len(pdf_reader.pages)
text = ""
for i in range(20, no_of_pages):
    page = pdf_reader.pages[i]
    text += page.extract_text()
```

```python

llama_model = AutoModelForCausalLM.from_pretrained(pretrained_model_name_or_path = "aboonaji/llama2finetune-v2",
                                                   quantization_config = BitsAndBytesConfig(load_in_4bit = True,
                                                                                            bnb_4bit_compute_dtype = getattr(torch, "float16"),
                                                                                            bnb_4bit_quant_type = "nf4"))
llama_model.config.use_cache = False
llama_model.config.pretraining_tp = 1
```

```python
llama_tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path = "aboonaji/llama2finetune-v2", trust_remote_code = True)
llama_tokenizer.pad_token = llama_tokenizer.eos_token
llama_tokenizer.padding_side = "right"
```

```python
from datasets import Dataset
```

```python
data = {"text": [text]}    # conversion of text into dictionary
dataset = Dataset.from_dict(data)


def tokenize_function(examples):
    return llama_tokenizer(
        examples["text"],
        padding="max_length",  # paddint to length
        truncation=True,     # if longer then reduced
        max_length=512,
        return_tensors="pt"  # return type supported in pytorch
    )

# converted into tokens
tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

```python
training_arguments = TrainingArguments(output_dir = "./results", per_device_train_batch_size = 4, max_steps = 100)
```

```python
llama_sft_trainer = SFTTrainer(
    model=llama_model,
    args=training_arguments,
    train_dataset=dataset,     # passing the dict of text
    tokenizer=llama_tokenizer,
    peft_config=LoraConfig(task_type="CAUSAL_LM", r=64, lora_alpha=16, lora_dropout=0.1),
    dataset_text_field="text"   # to focus on text field
)
```

```python

# Start training
llama_sft_trainer.train()
```

```python
user_prompt = "what is Quorum for meetings"
text_generation_pipeline = pipeline(
    task="text-generation",
    model=llama_model,
    tokenizer=llama_tokenizer,
    max_length=500
)


model_answer = text_generation_pipeline(user_prompt)
print(model_answer[0]['generated_text'])
```
